```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)

homophones = {'oh': '0', 'to': '2', 'three': '3','too':'2','for':'4'}

def replace_homophones(matcher, doc, match_id, matches):
    for match_id, start, end in matches:
        span = doc[start:end]
        new_tokens = []
        for token in span:
            if token.lower_ in homophones:
                new_tokens.append(homophones[token.lower_])
            else:
                new_tokens.append(token.text)
        new_doc = spacy.tokens.Doc(doc.vocab, words=new_tokens)
        span_start = span[0].idx
        span_end = span[-1].idx + len(span[-1])
        new_span = new_doc.char_span(span_start, span_end)
        if new_span is not None:
            doc.merge(new_span)





pattern = [{"LOWER": "temperature"}, {"POS": "AUX", "OP":"*"}, {"POS": "NUM"}, {"LOWER": "oh", "OP":"?"}, {"IS_PUNCT": True, "OP":"?"}, {"IS_DIGIT": True,},
        {"LOWER": "temperature"}, {"POS": "ADP", "OP":"*"}, {"POS": "NUM"}, {"IS_PUNCT": True, "OP":"?"},
        {"LOWER": "temperature"}, {"POS": "AUX", "OP":"*"}, {"IS_DIGIT": True}, {"IS_PUNCT": True, "OP":"?"},
        {"LOWER": "temperature"}, {"POS": "AUX", "OP":"*"}, {"POS": "NUM"}, {"LOWER": "oh", "OP":"?"}, {"IS_PUNCT": True, "OP":"?"}, {"POS": "NUM"}, {"IS_PUNCT": True, "OP":"?"},
        {"LOWER": "temp"}, {"POS": "AUX", "OP":"*"}, {"POS": "NUM"}, {"LOWER": "oh", "OP":"?"}, {"IS_PUNCT": True, "OP":"?"}, {"POS": "ADP", "OP":"*"}, {"IS_PUNCT": True, "OP":"?"}
          ]

matcher.add("Temperature", replace_homophones, pattern)

sentences = [
"Temperature was one Oh, 17 F. Heart rate was 108.",
"temperature of 100.8",
"No temperature 100.1",
"And as a temperature of 101.",
"Uh, temperature was 102.",
"Uh, his temperature was 100 degrees.",
"Um, temperature is 100.",
"Temperature is one Oh, one P 1 30.",
"Um, So, um um, he is on heartwarming Atlantic preventative and on physical exam Temperature is one oh, 2.3. Heart rate.",
"Her temperature is one Oh, 3.4 axillary.",
"Penny Sheehan Temp one oh, to heart rate 1 52.",
"Body temp was 11."
]
            

for sentence in sentences:
    doc = nlp(sentence)
    matches = matcher(doc)
    print(doc.text)
```

```python
# This code uses the spaCy library to process a text document containing patient numbers and find the digits from them. 
# It then removes any spaces in the digit sequences and joins the modified tokens back into a new string.
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

sentences = [   
"Temperature was 1 0 17 F. Heart rate was 108.",
"temperature of 100.8",
"No temperature 100.1",
"And as a temperature of 101.",
"Uh, temperature was 102.",
"Uh, his temperature was 100 degrees.",
"Um, temperature is 100.",
"Temperature is 1 0 1 P 1 30.",
"Um, So, um um, he is on heartwarming Atlantic preventative and on physical exam Temperature is 1 0 2.3. Heart rate.",
"Her temperature is 1 0 3.4 axillary.",
"Penny Sheehan Temp 1 0 2 heart rate 1 52.",
"Body temp was 11."
]

for sentence in sentences:
    doc = nlp(sentence)
    new_sentence = ""
    for token in doc:
        if token.text.isdigit():
            new_sentence += token.text
        else:
            new_sentence += token.text + " "
    print(new_sentence.strip() + "")
```
