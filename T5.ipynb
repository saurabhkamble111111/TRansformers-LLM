```python
import spacy
import pandas as pd
from happytransformer import HappyTextToText, TTSettings

# create a HappyTextToText object with T5 model and grammar correction fine-tuned model
happy_tt = HappyTextToText("T5", "vennify/t5-base-grammar-correction")

# set TTSettings for number of beams and minimum length of generated text
args = TTSettings(num_beams=5, min_length=1)

def correct_grammar(text):
    # Extract the sentences from the text
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    sentences = [sent.text for sent in doc.sents]

    # Correct the grammar of each sentence using the HappyTransformer T5 model
    corrected_sentences = []
    for sentence in sentences:
        prefixed_input_text = "grammar: " + sentence
        result = happy_tt.generate_text(prefixed_input_text, args=args)
        corrected_sentence = result.text.replace("grammar: ", "")
        corrected_sentences.append(corrected_sentence)

    # Create a list of tuples where the first element is the corrected sentence and the second element is the original sentence
    output = []
    for i in range(len(sentences)):
        if corrected_sentences[i] == sentences[i]:
            output.append((None, sentences[i]))
        else:
            output.append((corrected_sentences[i], sentences[i]))

    # Convert the output list to a pandas dataframe
    df = pd.DataFrame(output, columns=["Corrected Sentence", "Original Sentence"])

    # Save the dataframe to a CSV file
    df.to_csv("output.csv", index=False)

    return df
```
